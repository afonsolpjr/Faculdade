## Quest√£o 1: Calculando $\pi$ jogando dardos

### a) 
Suponha que num jogo de dardos voc√™ tem, pendurado em uma parede, o alvo (c√≠rculo de raio 1) devidamente emoldurado (inscrito em um quadrado de lado 2 - as unidades n√£o importam aqui). Por√©m, suas habilidades ao arremessar um dardo em dire√ß√£o ao alvo s√£o p√©ssimas, e voc√™ pode atingir qualquer ponto dentro do quadrado uniformemente (note que, for√ßas ocultas e inexplic√°veis do Universo fazem com que seja imposs√≠vel voc√™ acertar *fora* da moldura). Atrav√©s de um c√°lculo anal√≠tico mostre como que a probabilidade de voc√™ acertar o alvo se relaciona com o n√∫mero $\pi$.

#### Resolu√ß√£o item a)


Temos um quadrado de lado 2 (logo √°rea 4), com um c√≠rculo de raio 1 inscrito nele (logo √°rea =   $$\large \pi$$ )

De toda a √°rea poss√≠vel de ser acertada dentro do quadrado (nosso espa√ßo amostral), a √°rea que estamos interessada em acertar √© o c√≠rculo.

Como a √°rea do quadrado √© de tamanho 4, e a do c√≠rculo  √© $$\large \pi$$, se considerarmos lan√ßamentos aleat√≥rios, a chance de cair no c√≠rculo ser√° a raz√£o de sua √°rea com a do quadrado. 

Isto √©, $$ \frac{\large \pi}{4} \approx 0.78539816339 $$ 


### b) 
Utilizando gera√ß√£o de n√∫meros aleat√≥rios, fa√ßa uma simula√ß√£o para estimar o valor de $\pi$ atrav√©s do procedimento acima. Ou seja, simule que dardos s√£o arremessados uniformemente no quadrado e utilize a interpreta√ß√£o frequentista da probabilidade para aproximar o valor de $\pi$. Note que voc√™ precisar√° de "muitas" tentativas para ter uma boa estimativa. O qu√£o grande voc√™ acha que "muitas" deve ser?

```python
import numpy as np
import matplotlib.pyplot as plt

# Seja um c√≠rculo centrado no ponto (0,0)  e com raio = 1 o c√≠rculo que queremos acertar, 
# que est√° inscrito em um quadrado formado pelos pontos: (-1,-1) (-1,1) (1,1) (1,-1)

# Dado um ponto (x,y) aleat√≥rio,
# ele estar√° no alvo (ou dentro do c√≠rculo), se a sua dist√¢ncia at√© o centro do c√≠rculo for <= 1.

# Assim, sortearemos n pontos p1 dentro de 1/4 do espa√ßo do quadrado, no primeiro quadrante  ( 0 <= x, y <= 1 ) e checaremos
# se sua distancia at√© o centro √© menor ou igual a 1, e isto contar√° como um acerto.
# Note que embora estejamos falando s√≥ de um quadrante, as chances de acerto dentro da √°rea do quadrado s√£o as mesmas.


# Como a probabilidade neste sistema ser√° "acertos/n", que deve ser igual a "pi/4", logo temos pi = 4*acertos/n.

# Testaremos para v√°rios valores de n, tiraremos o erro absoluto considerando a constante pi da biblioteca numpy,
# e plotaremos o erro por n.


#valores de n
n_array = [2**i for i in range(22)]
#guardar√° valores de pi achado para os valores de n acima
pi_achado = []

for n in n_array:
    acertos = 0
    for i in range(n):
        x = np.random.random()
        y = np.random.random()
        distancia = np.sqrt( x*x + y*y )
        if(distancia<=1):
            acertos+=1
    pi_achado.append( 4*acertos/n )
    
#computando erros absolutos
erros_abs  = [ np.absolute( np.pi - x) for x in pi_achado ]


plt.plot(range(22),erros_abs,"k.", label="Erro absoluto",linestyle="dotted")
plt.title("Erro absoluto por n¬∫ de tentativas")
plt.xlabel("log2(n¬∫ de tentativas) ")
plt.ylabel("Erro absoluto")
plt.legend(fontsize=10)
plt.grid()
plt.show()

print("Temos para aproxidamente 2 milh√µes de tentativas o erro absoluto de {}".format(erros_abs[-1]))
```

## Quest√£o 2: O problema de Monty Hall

Em um famoso programa de televis√£o da d√©cada de 70 nos Estados Unidos, Monty Hall, o apresentador, apresentava tr√™s portas aos concorrentes. Atr√°s de uma delas estava um pr√™mio e, atr√°s das outras duas n√£o havia nada. A intera√ß√£o se dava em tr√™s etapas:

* Na 1a. etapa o concorrente escolhe uma das tr√™s portas (que ainda n√£o √© aberta);
* Na 2a. etapa, Monty abre uma das outras duas portas que o concorrente n√£o escolheu, revelando que o pr√™mio n√£o se encontra nessa porta e revelando uma porta vazia (note que o apresentador **sempre** ir√° abrir uma porta que est√° vazia, e **nunca** a porta com o pr√™mio);
* Na 3a. etapa Monty pergunta ao concorrente se quer decidir permanecer com a porta que escolheu no in√≠cio do jogo ou se ele pretende mudar para a outra porta que ainda est√° fechada para ent√£o a abrir. Agora, com duas portas apenas para escolher ‚Äî pois uma delas j√° se viu, na 2a. etapa, que n√£o tinha o pr√™mio ‚Äî e sabendo que o pr√™mio est√° atr√°s de uma das restantes duas, o concorrente tem que tomar a decis√£o.

O objetivo dessa quest√£o √© estudar qual √© a estrat√©gia mais l√≥gica: ficar com a porta escolhida inicialmente ou mudar de porta? Do ponto de vista probabl√≠stico, com qual das duas portas ainda fechadas o concorrente tem mais probabilidades de ganhar?

### a) 
Primeiramente, pensando do ponto de vista do participante, argumente se √© mais razo√°vel que ele tome sua decis√£o pensando do ponto de vista da interpreta√ß√£o frequentista ou subjetiva do conceito de probabilidade. Claro, justifique sua escolha!

Obs.: Esse item √© engra√ßado pois uma vez que o c√°lculo da probabilidade independe da interpreta√ß√£o escolhida, qualquer resposta √© v√°lida, do ponto de vista formal. Por√©m, a gra√ßa aqui √© *como* voc√™ vai fazer a sua justificativa para a interpreta√ß√£o escolhida!

#### Resolu√ß√£o item a)

Subjetivamente, a interpreta√ß√£o mais intuitiva que eu tive foi:
Dado que, ap√≥s a segunda etapa, somente uma das duas portas tem o premio, tanto faz escolher uma ou outra. A chance √© de 50% para manter ou mudar de porta!

Do ponto de vista frequentista, a chance de eu acertar a primeira porta dado v√°rias tentativas ser√° de 1/3. Enquanto que, para a outra porta que eu posso escolher ap√≥s a 2¬™ etapa, a chance √© de 50%. Logo √© mais razo√°vel escolher mudar de porta.

### b)
Talvez voc√™ j√° saiba a resposta para esse problema, uma vez que ele √© bem famoso, mas n√£o tem problema! Fa√ßa simula√ß√µes para encontrar qual √© o melhor cen√°rio para o participante: manter sua escolha inicial ou mudar de porta. Justifique, **intuitivamente** (ou seja, **sem fazer contas**) porque essa escolha √© razo√°vel.

```python
import numpy as np
import matplotlib.pyplot as plt


# Fun√ß√£o recebe array com n tentativas
def jogo(n, mudar_porta):
    acertos = 0
    for i in range(n):
        portas = [False, False, False]
        # escolhendo aleatoriamente porta com premio
        portas[np.random.randint(3)] = True
        # fazendo 1¬™ escolha
        escolha = portas[np.random.randint(3)]
        # etapa da 2¬™ escolha
        if(mudar_porta and not escolha):
            acertos+=1
        if(not mudar_porta and escolha):
            acertos+=1
    return acertos
                


n = [2**i for i in range(21)]

prob_acertos_mudando = [ jogo(x,True)/x for x in n ]

prob_acertos_mantendo = [ jogo(x,False)/x for x in n ]


plt.plot(range(21),prob_acertos_mantendo,"r.", label="Prob. caso mantenha" ,linestyle="dotted")
plt.plot(range(21),prob_acertos_mudando, "b.", label="Prob. caso mude" ,linestyle="dotted")
plt.legend(fontsize=10)
plt.grid()
plt.title("Prob. de acerto por tentativas por escolha")
plt.xlabel("log2(n¬∫ tentativas)")
plt.ylabel("Probabilidade de acertar")
plt.show()


print("Vemos que a probabilidade de acertar caso mude converge para 2/3. Enquando a de acertar caso mantenha converge para 1/3")

    
    
    
```

### Resolu√ß√£o item c)

Seja:
- A o evento "ganhar sem trocar a porta"
- B o evento "ganhar trocando a porta"
- C o evento "acertar a porta de primeira"
- D o evento "errar a porta de primeira".

Temos que:
- P(C) = 1/3
- P(D) = 2/3

Como C e D s√£o disjuntos, temos que:
- P(A) = P(A|C) + P(A|D)
- P(B) = P(B|C) + P(B|D)

Observando que ganhar sem trocar a porta s√≥ √© poss√≠vel se voc√™ acertar de primeira, e que ganhar trocando a porta s√≥ √© poss√≠vel se voc·∫Ω errar de primeira,temos:

- P(A)=P(A|C) = 1/3
- P(B)=P(B|D) = 2/3

## Quest√£o 3: Jogo da velha aleat√≥rio

Agora, vamos tentar estimar uma probabilidade cujo c√°lculo anal√≠tico √© bem mais complexo. Suponha que dois jogadores joguem jogo da velha por√©m com uma estrat√©gia um pouco ex√≥tica: cada jogador escolhe a casa onde ir√° marcar $\times$ ou $\circ$ uniformemente dentre as casas vazias. Exemplificando, o primeiro jogador marca $\times$ em uma das 9 casas com igual probabilidade; o segundo marca $\circ$ em uma das 8 casas restantes com igual probabilidade; e assim sucessivamente at√© algum jogador ganhar (linha, coluna ou diagonal completa com $\times$ ou $\circ$) ou empatar (todas as c√©lulas preenchidas por√©m sem nenhuma linha nem coluna nem diagonal completa com $\times$ ou $\circ$).

### a) 
Encontre uma aproxima√ß√£o para a probabilidade do jogo **n√£o** terminar em empate. Tente encontrar uma justificativa intuitiva para sua resposta.

#### Resolucao do item a se encontra junto √† resolu√ß√£o do item b.

### b) 
Dentre os jogos que n√£o terminam empatados, qual jogador tem mais chances de ganhar, se ambos utilizam essa estrat√©gia: o jogador que come√ßa ou o segundo jogador? Tente encontrar uma justificativa intuitiva para sua resposta.

```python
# Fazendo as simula√ß√µes para o jogo da velha normal.

# determinando conjuntos ganhadores:
vitorias = [{0,1,2},{3,4,5},{6,7,8},{0,3,6},{1,4,7},{2,5,8},{0,4,8},{2,4,6} ]


#retorna vencedor do jogo(1,2) ou 0 caso empate
def jogo_velha(vitorias):
    casas_disponiveis = [i for i in range(0,9)]
    p1 = set()
    p2 = set()
    turno = 0
    while( turno < 9 ):
        casa_escolhida = np.random.choice(casas_disponiveis) # escolhe numero entre disponiveis
        if not turno%2:                               #adiciona numero escolhido ao conjunto do jogador atual
            p1.add(casa_escolhida)
        else:
            p2.add(casa_escolhida)
        casas_disponiveis.remove(casa_escolhida)             #remove numero escolhido

        if(turno>=4):              #checar se alguem venceu e retornar o numero do vencedor
            for conjunto in vitorias:
                if(not turno%2):
                    if (conjunto.issubset(p1)):
                        return 1
                elif(conjunto.issubset(p2)):
                    return 2
        turno+=1
    return 0            #caso empate
        
#determinando numero de tentativas                    
n_tentativas = [2**i for i in range(17)]

#computando resultados
resultados = []
for n in n_tentativas:
    amostras = [ jogo_velha(vitorias) for i in range(n)] #retorna lista com n entradas, uma para cada jogo, contendo resultado
    resultados.append(amostras)
    
#contando resultados
empates = [ jogo.count(0) for jogo in resultados ]
vit_p1 = [ jogo.count(1) for jogo in resultados ]
vit_p2 = [jogo.count(2) for jogo in resultados ]

#calculando probabilidades
prob_empate = [ x/y for x,y in zip(empates,n_tentativas) ]
prob_p1 = [ x/y for x,y in zip(vit_p1,n_tentativas) ]
prob_p2 = [ x/y for x,y in zip(vit_p2,n_tentativas) ]

#plotando graficos
plt.title("Probabilidade do resultado por n¬∫ de tentativas")
plt.grid()
plt.xlabel("Log2(n¬∫ tentativas)")
plt.ylabel("Probabilidade do resultado")
plt.plot(range(17),prob_p1,"bo",linestyle="dotted",label = "prob. vit√≥ria jogador 1" )
plt.plot(range(17),prob_p2,"ro",linestyle="dotted",label = "prob. vit√≥ria jogador 2" )
plt.plot(range(17),prob_empate,"ko",linestyle="dotted",label = "prob. empate" )
plt.legend(fontsize=10)
plt.show()

print("A probabilidade do jogador 1 vencer converge para aproximadamente {:.2f}%\n".format((prob_p1[-1]+prob_p1[-2])/2 *100) +
      "A probabilidade do jogador 2 vencer converge para aproximadamente {:.2f}%\n".format((prob_p2[-1]+prob_p2[-2])/2 *100) +
      "A probabilidade de empate converge para aproximadamente {:.2f}%".format((prob_empate[-1]+prob_empate[-2])/2 *100) )
      
      
      
```
